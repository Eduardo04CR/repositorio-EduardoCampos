<h4>2.¿Que tipo de propiedad tiene la mayor facilidad de venta?<h/4>

```python
gf2 = df_main.groupby(["propertyType"]).count()
gf2 = gf2.iloc[:,:1]
gf2 = gf2.sort_values("serialNumber", ascending=True)
gf2
```

```python
fig, ax = plt.subplots(1,1, figsize =(15,6.9), dpi=100)

propiedades = gf2.index.tolist()
cantidad = gf2["serialNumber"].tolist()

bar_colors = ['#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
            '#e5e5e5','#e5e5e5',"#e5e5e5","#e5e5e5",'#b19deb']

ax.barh(propiedades, cantidad, color=bar_colors)

#--------------------------------------------------------------------------------------------#
#-----------------CONFIGURACION DE LOS TITULOS, SUBTITULOS Y EFES DEL GRAFICO----------------#
#--------------------------------------------------------------------------------------------#
#FORMATO DEL TITULO

title = "Grafico 1: Tipos de Propiedades"
plt.suptitle(title,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.87,
             transform = fig.transFigure,
             fontsize = 20,
             fontweight = "bold",
             fontfamily = "serif",
             fontname = "Liberation Sans Narrow")

#FORMATO DEL SUBTITULO
subTitle = "Cantidad de propiedades que tienen mayor facilidad de venta"
ax.set_title(subTitle,
             horizontalalignment = "left",
             x = 0.001,
             y = 0.80,
             transform = fig.transFigure,
             fontsize = 15,
             fontweight = "regular",
             fontfamily = "serif")

#FORMATO AL EJE (X/Y)
etiqueta = "Cantidad"
ax.set_xlabel(etiqueta,
              ha = 'left',
              fontsize = 10,
              fontweight = "regular",
              fontfamily = "serif")

ax.xaxis.set_label_coords(x=0, y=-0.07, transform = ax.transAxes)
etiquetay = "Tipo de propiedades"
ax.set_ylabel(etiquetay,
              ha = 'left',
              fontsize = 10,
              fontweight = "regular",
              fontfamily = "serif")

ax.xaxis.set_label_coords(x=0, y=-0.07, transform = ax.transAxes)

#FORMATO TAMANO LETRA AMBOS EJES
ax.tick_params(axis='both', labelsize = 10)

#CONFIGURA LA UBICACION EL TITULO, SUBTITULO Y GRAFICO
plt.subplots_adjust(top = 0.80)

plt.show()
```

<p>El tipo de propieda que tiene mayor de facilidad de venta es el detipo "Residencial", ya que alcanza más de 60 mil registros de venta</p>
