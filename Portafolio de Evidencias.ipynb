<h4 Style = " color:#A0522D"> 2.3 Imputación Simple de Valores Numericos <h/4>

```python
proNormalizadas = dfAutos ["normalized"].astype("float").mean(axis=0)  # para sacar el promedio
dfAutos["normalized"].replace(np.nan,proNormalizadas,inplace=True) #aqui decios que reemplace 
dfAutos.head()                                                     #los datos faltantes en la columan 
                                                                #por el promedio que sacamos
```

```python
probore = dfAutos ["bore"].astype("float").mean(axis=0)
dfAutos["bore"].replace(np.nan,probore,inplace=True)
dfAutos.head()
```

```python
prostroke = dfAutos ["stroke"].astype("float").mean(axis=0)
dfAutos["stroke"].replace(np.nan,prostroke,inplace=True)
dfAutos.head()
```

```python
prohorsepower = dfAutos ["horsepower"].astype("float").mean(axis=0)
dfAutos["horsepower"].replace(np.nan,prohorsepower,inplace=True)
dfAutos.head()
```

```python
propeak = dfAutos ["peak"].astype("float").mean(axis=0)
dfAutos["peak"].replace(np.nan,propeak,inplace=True)
dfAutos.head()
```

<h4 Style = " color:#A0522D"> 2.4 Imputación Simple de Valores categoricos <h/4>

```python
dfAutos["doors"].value_counts().idxmax() # buscar el que mas se repite, para imputar valores tipo caracter
```

```python
dfAutos["doors"].replace(np.nan,"four",inplace=True) # reemplazamos los nan por el valor repetido encontrado
```

<h4 Style = " color:#A0522D"> 2.5 Eliminar filas con valores faltantes <h/4>

```python
dfAutos.dropna(subset = ["price"], axis=0, inplace=True) #para borrar columna con valor faltante
dfAutos.reset_index (drop=True, inplace=True)   # reinicia el indice
```

```python
dfTemp = dfAutos.isnull()  # verificación de nuevo para ver si todos los datos faltantes fueron trabajados
for columna in dfTemp.columns.values.tolist() :
    print(columna)
    print(dfTemp[columna].value_counts())
    print("")
```

<h4 Style = " color:#A0522D"> 3.1 Verificar el dominio de cada una de las variables <h/4>

```python
dfAutos.dtypes
```

```python
dfAutos[["normalized","horsepower","peak","price"]] = dfAutos[["normalized","horsepower","peak","price"]].astype("int") 
#para pasar los datos a variables enteras

dfAutos[["bore","stroke"]] = dfAutos[["bore","stroke"]].astype("float")
#para pasar los datos a variables con coma
```

```python
dfAutos.dtypes
```

<h4 Style = " color:#A0522D"> 3.2 Estandarizar los datos <h/4>

dfAutos.head(3)

```python
dfAutos["City L/100km (calc)"] = 235/dfAutos["citye"] # se crea una nueva columna para topicalizar los datos
dfAutos["Higway L/100km (calc)"] = 235/dfAutos["higway"] #aqui estamos pasando millas por galon a kilometros por litro
```

```python
dfAutos.head(3)
```

<h4 Style = " color:#A0522D"> 3.3 Volver variables comparables <h/4>

```python
dfTemp = dfAutos[["horsepower","peak"]]
dfTemp.head(3)
```

```python
dfAutos["horsepower (norm)"] = dfAutos["horsepower"] / dfAutos["horsepower"].max() # para volver valores comparables si los valores están muy desequilibrados
dfAutos["peak (norm)"] = dfAutos["peak"] / dfAutos["peak"].max() # Casi no se utiliza al menos si solictan una comparación entre dos casos diferentes
dfAutos.head(5)
```

<h4 Style = " color:#A0522D"> 3.4 Bining <h/4>

```python
dfTemp.head(5)
```

```python
bins =np.linspace (48,262,4)
bins
```

```python
categoria = ["bajo","medio","alto"]
dfAutos["horsepower (bins)"] = pd.cut(dfAutos["horsepower"],bins,labels = categoria,include_lowest=True)
dfAutos[["horsepower","horsepower (bins)"]].head(5)
```

```python
#Remueve las lineas de los axis, en los graficos 
mpl.rcParams["axes.spines.bottom"] = True
mpl.rcParams["axes.spines.left"] = False
mpl.rcParams["axes.spines.right"] = False
mpl.rcParams["axes.spines.top"] = False
```

```python
gf = dfAutos.groupby(['make']).count()
gf = gf.iloc[:,:1]
gf = gf.sort_values('symboling', ascending=False)
gf
```

```python
fig, ax = plt.subplots(1,1, figsize =(15,6.9), dpi=100)

eje_x = gf.index.tolist()
data = gf["symboling"].tolist()

bar_colors = ['#69d2e7','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
             '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
             '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#fa6900']

ax.barh(eje_x, data, color=bar_colors)

#--------------------------------------------------------------------------------------------#
#-----------------CONFIGURACION DE LOS TITULOS, SUBTITULOS Y EFES DEL GRAFICO----------------#
#--------------------------------------------------------------------------------------------#
#FORMATO DEL TITULO

title = "Grafico 1: Vehiculos en Inventario"
plt.suptitle(title,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.87,
             transform = fig.transFigure,
             fontsize = 20,
             fontweight = "bold",
             fontfamily = "serif",
             fontname = "Liberation Sans Narrow")

#FORMATO DEL SUBTITULO
subTitle = "Cantidad de vehiculos agrupados segun fabricante, en Costa Rica, periodo 2023"
ax.set_title(subTitle,
             horizontalalignment = "left",
             x = 0.001,
             y = 0.80,
             transform = fig.transFigure,
             fontsize = 15,
             fontweight = "regular",
             fontfamily = "serif")

#FORMATO AL EJE (X/Y)
etiqueta = "cantidad de carros (ABS)"
ax.set_xlabel(etiqueta,
              ha = 'left',
              fontsize = 10,
              fontweight = "regular",
              fontfamily = "serif")

ax.xaxis.set_label_coords(x=0, y=-0.07, transform = ax.transAxes)

#FORMATO TAMANO LETRA AMBOS EJES
ax.tick_params(axis='both', labelsize = 10)

#CONFIGURA LA UBICACION EL TITULO, SUBTITULO Y GRAFICO
plt.subplots_adjust(top = 0.80)

plt.show()
```
